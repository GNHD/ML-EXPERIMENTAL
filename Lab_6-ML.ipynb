### 2.

```python
def entropy(target_col):

    elements,counts=np.unique(target_col,return_counts=True)
    entropy=np.sum([(-counts[i]/np.sum(counts))*np.log2(counts[i]/np.sum(counts)) for i in range(len(elements))])
    return entropy
```

### 3.

```python
uniqueValues=data['vhigh'].unique()
print(uniqueValues)
```

```python
def InfoGain(data,split_attribute_name,target_name="class"):
    total_entropy = entropy(data[target_name])
    vals,counts=np.unique(data[split_attribute_name],return_counts=True)
    Weighted_Entropy = np.sum([(counts[i]/np.sum(counts))*entropy(data.where(data[split_attribute_name]==vals[i]).dropna()[target_name]) for i in range(len(vals))])
    Information_Gain = total_entropy-Weighted_Entropy
    return Information_Gain
```

### 4.

```python
print(InfoGain(data,'vhigh','unacc'))
print(InfoGain(data,'vhigh.1','unacc'))
print(InfoGain(data,'2','unacc'))
print(InfoGain(data,'2.1','unacc'))
print(InfoGain(data,'small','unacc'))
print(InfoGain(data,'low','unacc'))
```
